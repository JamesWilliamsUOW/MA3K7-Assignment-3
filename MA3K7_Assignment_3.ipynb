```python
def hat_finals(n):
    memo = {}

    def explore(state):
        state = tuple(sorted(state))
        
        if state in memo:
            return memo[state]
        
        if len(state) == 1:
            return {state[0]}
        
        finals = set()

        for i in range(len(state)):
            for j in range(i + 1, len(state)):
                a = state[i]
                b = state[j]
                diff = abs(a - b)

                new_state = list(state)
                new_state.pop(j)
                new_state.pop(i)
                new_state.append(diff)

                results = explore(new_state)
                finals = finals.union(results)
        
        memo[state] = finals
        return finals

    start = list(range(1, n + 1))
    return explore(start)
```

```python
import random
import matplotlib.pyplot as plt

def simulate_hat(n, trials):
    results = []
    
    for t in range(trials):
        # Start with numbers 1 to n
        hat = list(range(1, n + 1))
        
        # Keep going until one number remains
        while len(hat) > 1:
            # Choose two different random positions
            i = random.randint(0, len(hat) - 1)
            j = random.randint(0, len(hat) - 1)
            
            # Make sure they are different
            while j == i:
                j = random.randint(0, len(hat) - 1)
            
            a = hat[i]
            b = hat[j]
            diff = abs(a - b)
            
            # Remove larger index first
            if i > j:
                hat.pop(i)
                hat.pop(j)
            else:
                hat.pop(j)
                hat.pop(i)
            
            # Add the difference
            hat.append(diff)
        
        # Store final value
        results.append(hat[0])
    
    # Plot histogram
    plt.hist(results, bins=range(0, n + 2), edgecolor='black')
    plt.xlabel("Final Value")
    plt.ylabel("Frequency")
    plt.title(f"Hat Simulation: n = {n}, trials = {trials}")
    plt.savefig(f"histogram_n{n}.png", dpi=300, bbox_inches="tight")
    plt.show()
    
    return results
```
